```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

### Basic graph

```python
#ploting the points
x=[0,1,2,3,4]
y=[0,2,4,6,8]

#resize the graph (dpi means pixels per inch, use 300 when saving
plt.figure(figsize=(5,3), dpi=200)

#keyword argument notation
#plt.plot(x,y, label='2X',color='red', linewidth=3, marker='.', linestyle='--', markersize=10, markeredgecolor='blue')

#using shorthand notation
#fnt = '(color)(marker)(line)'
plt.plot(x,y, 'b','--', label='2x')

#another line
#select intervals for te points
x2 = np.arange(0,4.5,0.5)

#plot part of the graph as line
plt.plot(x2[:6], x2[:6]**2, 'r', label='x^2')

#plot remaining as dots
plt.plot(x2[5:], x2[5:]**2, 'r--')

#add title and specify font parameters with fontdict
plt.title('My First Grath', fontdict={'fontname':'Comic Sans MS', 'fontsize':15})

#x and y labels
plt.xlabel('x-axis')
plt.ylabel('y-axis')

#x y axis thickness
plt.xticks([0,1,2,3,4])
#plt.yticks([0,2,4,6,8,10])

#add legend
plt.legend()

#save figure dpi of 300 is good so that graph has a high resolution
plt.savefig('mygraph.png', dpi=300)

#show graph
plt.show
```

### Bar chart

```python
labels = ['A', 'B', 'C']
values = [1,4,2]

bars = plt.bar(labels, values)

patterns = ['/', 'o', '*']
for bar in bars:
    bar.set_hatch(patterns.pop())

#bar[0].set_hatch('/')
#bar[1].set_hatch('o')
#bar[2].set_hatch('*')

plt.figure(figsize=(6,4))

plt.show
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
sns.set()
```

```python
fifa = pd.read_csv(r"C:\Users\LENOVO\Documents\data bank\fifa19.csv")
```

```python
fifa.head()
```

```python
#Creating histogram
plt.hist(fifa['Overall'])
plt.show()
```

```python
#Scatter Plot
plt.scatter(fifa['Overall'], fifa['Wage'])
plt.title('Overall Vs. Wages')
plt.ylabel('Wages')
plt.xlabel('Overall')


plt.show()
```

```python
# Bar Graph

from collections import Counter
#print(Counter(fifa['Nationality']))

nationality_dict = dict(Counter(fifa['Nationality']).most_common(10))
nationality_dict
```

```python


plt.bar(nationality_dict.keys(), nationality_dict.values(), color='blue')
plt.xlabel('Nationality')
#color = Nationality
plt.ylabel('Frequency')
plt.title('Bar Plot of 10 Most Common Nationalities')
plt.xticks(rotation=45)
plt.show()
```

```python
#pie chart

plt.figure(figsize=(8,8))
plt.pie(nationality_dict.values(), labels=nationality_dict.keys(), autopct='%1.2f%%')
plt.axis('equal')

plt.show()
```

## Generating Heatmaps 

```python
corr = fifa[['Overall', 'Age', 'Wage', 'Skill Moves']].corr()
corr
```

```python
#plt.figure(figsize=(6,6))
sns.heatmap(corr, annot=True)
plt.title('Heatmap of Overall, Age, Wage and Skill Moves')

plt.show()
```

# Pair Plot

```python
pair = fifa[['Overall', 'Age', 'Potential']]
sns.pairplot(pair)
plt.title('Pair Plot of Overall, Age and Potential')

plt.show()
```

```python
#pcolor()

import matplotlib.pyplot as plt
import numpy as np
```

```python
#creating a nonsquare matrix
data = np.random.rand(4,6)
plt.pcolor(data)
plt.show()
```

```python
#custom the color
data = np.random.rand(4,6)
plt.pcolor(data, cmap='BrBG')
plt.show()
```

```python
# adding shading
N=200
X, Y = np.meshgrid(np.linspace(-5,5,N), np.linspace(-4,4,N))
Z = (X**2 + Y**3)

#visualising
plt.pcolor(X,Y,Z, cmap='YlGnBu')
plt.show()
```

```python
#create container dictioary
avg{}
#split data into different regions
for make in auto_dt['make'].unique():
    tempauto_dt = auto_dt[auto_dt['make']==make]
    
    #apply agregation function
    average = tempauto_dt['price'].mean()
    #combine the data into dataframe
    averages[make] = [average]

aggregate_df = pd.DataFrame.from_dict(averages, orient='index', columns=['Average Price'])
print(aggregate_df)
```

# using your own function

```python
#function name is group_range

def group_range(x):
    return x.max() - x.min()
ranges = auto_dt.groupby(['make', 'fuel_type'])['price'].apply(group_range)
ranges
```

```python
ranges = auto_dt.groupby(['make', 'fuel_type']).apply(lambda x: x['price'].max()<20000)
ranges
```

```python
# filtering using group by
ranges = auto_dt.groupby(['make', 'fuel_type']).filter(lambda x: x['price'].max()<20000)
ranges
```

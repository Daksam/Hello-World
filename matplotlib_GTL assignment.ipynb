```python
# Filter the data to select only the four countries and the Years column
df = gas[['Year', 'Canada','Germany','UK', 'USA']]

# Plotting
plt.figure(figsize=(10, 6))

# Plotting each country's data
for country in ['Canada','Germany','UK', 'USA']:
    plt.plot(df['Year'], df[country], label=country, marker='*', linestyle='-')


# Adding labels and title
plt.xlabel('Year')
plt.ylabel('Gas Price')
plt.title('Gas Prices for Four Selected Countries Over the Years')

# Adding legend
plt.legend()

# Display plot
plt.grid(True)
plt.show()

plt.savefig('SamuelDassi_line.png')
```

### 2. Pie plot activity
<br> Create a pie plot for the data.
- set colors
- explode one section
- indicate their percentages
- legend(give it a title eg Education level)

```python
workers=[40,35,100,70]
education=['Primary','JHS','SHS','Tertiary']
```

```python
explode = (0, 0, 0, 0.05)  # Explode the 'Tertiary' section

# Plotting
plt.figure(figsize=(6, 6))
colors = ['red', 'blue', 'yellow', 'green']  # Custom colors

# Plotting pie chart
plt.pie(workers, labels=education, colors=colors, autopct='%1.1f%%', startangle=140, explode=explode)

# Title
plt.title('Education Level of Workers')

# Legend
plt.legend(title='Education Level', loc='upper right')

# Display plot
plt.axis('equal')  # Equal aspect ratio ensures that pie is drawn as a circle.
plt.show()

plt.savefig('SamuelDassi_pie.png')
```

### 3. Scatter plot activity
Create a scatter plot with matplotlib that displays xt paired with yt and zt:

```python
xt=np.array([0,0.1,0.2,0.3,0.5,0.8,1.0])
yt=np.array([1.0,2.1,3.5,6.5,7.2,5.9,6.3])
zt= xt*yt
```

Change the shape of the markers to
1. square for yt
2. triangle for zt
<br> Also:
- Add a label to indicate which points are yt and zt.
- add title
- xlabel and ylabel

```python
# Plotting
plt.figure(figsize=(8, 6))

# Scatter plot for yt
plt.scatter(xt, yt, label='yt', color='blue', marker='s')

# Scatter plot for zt
plt.scatter(xt, zt, label='zt', color='red', marker='^')

# Adding labels and title
plt.xlabel('xt')
plt.ylabel('yt and zt')
plt.title('Scatter Plot of xt against yt and zt')
plt.legend()

# Display plot
plt.grid(True)
plt.show()

plt.savefig('SamuelDassi_scater.png')
```

### 4. Bar plot Activity
create a bar plot with
- color
- xlabel
- ylabel
- title



```python
course=['Matlab','R','Python','Excel','Java']
values=[20,15,57,60,30]
```

```python
# Sorting the data in descending order
sorted_values, sorted_courses = zip(*sorted(zip(values, course), reverse=True))

# Plotting
plt.figure(figsize=(8, 6))

# Bar plot
plt.bar(sorted_courses, sorted_values, color=['blue', 'green', 'red', 'purple', 'orange'])

# Adding labels and title
plt.xlabel('Courses')
plt.ylabel('Values')
plt.title('Bar Plot of Courses')

# Display plot
plt.show()

plt.savefig('SamuelDassi_bar.png')
```

### 5. Histogram Activity
Create a histogram of the data below
<br> modify it to your liking

```python
from random import randint
marks=[randint(20,90) for i in range(50)]
bins=[40,50,60,70,80,90,100]
```

```python
# Plotting
plt.figure(figsize=(8, 6))

# Histogram
plt.hist(marks, bins=bins, color='green', edgecolor='black')

# Adding labels and title
plt.xlabel('Marks')
plt.ylabel('Frequency')
plt.title('Histogram of Marks')

# Adding grid
plt.grid(True)

# Display plot
plt.show()

plt.savefig('SamuelDassi_histogram.png')
```

## 6. Heatmap
Create a heatmap using the correlation of the gas_prices 
* add a title

```python
# Calculate correlation matrix
correlation_matrix = gas.corr()

# Plot heatmap
plt.figure(figsize=(10, 6))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', fmt=".2f")

# Adding title
plt.title('Correlation Heatmap of Gas Prices')

# Display plot
plt.show()

plt.savefig('SamuelDassi_heatmap.png')
```

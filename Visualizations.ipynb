## Matplotlib.pyplot

### Plotting line chart


```python
plt.plot(iris.sepal_length)
plt.title('Line plot of Sepal Length')
```

```python
plt.figure(figsize=(8,5))
plt.plot(iris.sepal_length,iris.petal_length, color='b', marker='*', label='iris')
plt.xlabel('Sepal Length')
plt.ylabel('Petal Length')
plt.title('Line plot of Petal Length vs Sepal Length')
plt.legend()
plt.show()
```

### pie plot

```python
iris.species.value_counts()
```

```python
names = ['edi','sam','ray','seth']
marks = [80,120,66,33]
```

```python
plt.figure(figsize=(8,5))
plt.pie(marks,labels=names,
        autopct='%.2f%%',
        colors=['red','blue','green','yellow'],
        explode=[0,.1,0,0],
        shadow=True
       )
plt.title('general family data')
plt.legend(loc='lower center',ncol=4)


;
```

### Bar plots

```python
plt.figure(figsize=(8,5))
plt.bar(names, marks,
       color=['red','blue','green','yellow']
       )
plt.xlabel('Names')
plt.ylabel('Marks')
plt.title('general family data')

;
```

```python
plt.figure(figsize=(8,5))
plt.barh(names, marks,
       color=['red','blue','green','yellow']
       )

for index, value in enumerate(marks):
    plt.text(value, index,
             str(value))
    
plt.xlabel('Names')
plt.ylabel('Marks')
plt.title('general family data')

;
```

### Scatter plot

```python
plt.figure(figsize=(8,5))

plt.scatter(iris.sepal_length,iris.petal_length,
            color='b',
            marker='.',
            label='iris')

plt.xlabel('Sepal Length')
plt.ylabel('Petal Length')
plt.title('Scatter plot of Petal Length vs Sepal Length')
plt.legend()
plt.show()
```

### Subplots

```python
plt.figure(figsize=(10,10))

plt.subplot(2,2,1)
plt.barh(names, marks,
       color=['red','blue','green','yellow']
       )

for index, value in enumerate(marks):
    plt.text(value, index,
             str(value))
    
plt.xlabel('Names')
plt.ylabel('Marks')
plt.title('general family data')


plt.subplot(2,2,2)
plt.pie(marks,labels=names,
        autopct='%.2f%%',
        colors=['red','blue','green','yellow'],
        explode=[0,.1,0,0],
        shadow=True
       )
plt.title('general family data')
plt.legend(loc='lower center',ncol=4)


plt.subplot(2,2,3)
plt.scatter(iris.sepal_length,iris.petal_length,
            color='b',
            marker='.',
            label='iris')

plt.xlabel('Sepal Length')
plt.ylabel('Petal Length')
plt.title('Scatter plot of Petal Length vs Sepal Length')
plt.legend()


plt.subplot(2,2,4)
plt.plot(iris.sepal_length)
plt.title('Line plot of Sepal Length')

plt.suptitle('Dashboard for our plots', fontsize=16)

plt.savefig('dashboard.png')
;

```

```python
plt.figure(figsize=(8,15))
plt.subplot(4,1,1)

plt.barh(names, marks,
       color=['red','blue','green','yellow']
       )

for index, value in enumerate(marks):
    plt.text(value, index,
             str(value))
    
plt.xlabel('Names')
plt.ylabel('Marks')
plt.title('general family data')


plt.subplot(4,1,2)
plt.pie(marks,labels=names,
        autopct='%.2f%%',
        colors=['red','blue','green','yellow'],
        explode=[0,.1,0,0],
        shadow=True
       )
plt.title('general family data')
plt.legend(loc='lower center',ncol=4)


plt.subplot(4,1,3)
plt.scatter(iris.sepal_length,iris.petal_length,
            color='b',
            marker='.',
            label='iris')

plt.xlabel('Sepal Length')
plt.ylabel('Petal Length')
plt.title('Scatter plot of Petal Length vs Sepal Length')
plt.legend()


plt.subplot(4,1,4)
plt.plot(iris.sepal_length)
plt.title('Line plot of Sepal Length')

plt.suptitle('Dashboard for our plots', fontsize=16)

plt.tight_layout()
;
```
